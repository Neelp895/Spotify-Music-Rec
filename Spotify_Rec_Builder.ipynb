```python
#function drops non-numeric data and sums the properties 
def playlistvector(data):
    data = data.drop(columns= ['track_id', 'name'])
    data = data.sum()
    return data
```

```python
playlist_vec = playlistvector(user_playlist_df)
```

```python
print(playlist_vec)
```

```python
## converting to numpy array
np_playlist = playlist_vec.values
print(np_playlist)
```

```python
## function computes cos similarity
def cos_sim(row, v1):
    v2 = row.values
    result = np.dot(v1, v2)
    result = result/(np.linalg.norm(v1) * np.linalg.norm(v2))
    return result
```

```python
## function computes cos similarity
def dummy(row, a_val):
    v2 = row.values
    return v2[0]
```

```python
cos_similarities = df_subset.apply(cos_sim, axis = 1, args = (np_playlist,))
```
